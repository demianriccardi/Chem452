## Visualizing a viral capsid
A viral capsid is the protein shell that contains all of its evil good stuff.  In this notebook we are going to use the 'asymmetric unit' to create the coordinates of the entire capsid. The first reasonable one I found was for Human Hepatitis B (pdbid: 1QGT)... sounds like fun to me! To do this, we will use the symmetry operators that are available in the pdb file. Actually, before reading the rest of this notebook, you should load up this page:

http://www.rcsb.org/pdb/files/1QGT.pdb

and see if you can find the information we will use.

#### Next, we transform each set of coordinates and store them!
For fun, let's create Hg atom at the center of mass for the asymmetric unit.  This will give us a very coarse representation of the molecule.

```perl
my $hg = HackaMol::Atom->new(symbol => 'Hg', coords =>[$mol->COM] );
my $hg_ball = HackaMol::Molecule->new(name => 'Hg Ball') ;

foreach my $symop (keys %sym_op){
    my @mat_d = @{$sym_op{$symop}};
    my $cx = V(map{$_->[0]} @mat_d); 
    my $cy = V(map{$_->[1]} @mat_d);
    my $cz = V(map{$_->[2]} @mat_d);
    my $dxyz = V(map{$_->[3]} @mat_d);
# hg
    my $COM = $mol->COM; # center of mass
    my $COM_transform = $COM->[0]*$cx + $COM->[1]*$cy + $COM->[2]*$cz + $dxyz;
    $hg_ball->push_atoms( HackaMol::Atom->new(symbol => 'Hg', coords =>[$COM_transform]));
# the protein
    foreach my $atom ($mol->all_atoms){
        my ($x,$y,$z) = @{$atom->xyz};
        my $xyz_new = $x*$cx + $y*$cy + $z*$cz + $dxyz;  
        $atom->push_coords($xyz_new);
    }    
}
```
